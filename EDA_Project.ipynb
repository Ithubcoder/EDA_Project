**Exploratory Data Analysis using Python**

```python
# Importing Necessary Libraries.
import io
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Uploading Dataset / File.
from google.colab import files
uploaded = files.upload()
```

```python
# Reading the CSV File.
df = pd.read_csv(io.BytesIO(uploaded['data.csv']))
```

```python
# Cleaning the dataset.
company = df.drop(['Website','Description'], axis=1)
```

```python
company.head()
```

```python
# Corelation Matrix.
corelation = company.corr()
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns, annot=True)
```

```python
# Pair Plot.
sns.pairplot(company)
```

```python
# Relation Plot / Scatter Plot.
sns.relplot(x = 'Index', y = 'Founded', hue = 'Number of employees', data = company)
```

```python
# Histogram Plot.
plt.figure(figsize=(12, 5))
sns.distplot(company['Number of employees'])
```

```python
# Categorical Plot.
sns.catplot(x = 'Number of employees', kind = 'box', data = company)
```

```python
# Line Plot.
plt.figure(figsize=(28, 8))
sns.lineplot(data = company, x = "Founded", y = "Index", palette = "muted")
```

```python
# Bar Plot.
plt.figure(figsize=(28, 8))
sns.barplot(data = company, x = "Founded", y = "Number of employees", palette = "muted")
```

```python
# Strip Plot.
plt.figure(figsize=(28, 8))
sns.stripplot(data = company, x = "Founded", y = "Number of employees", dodge = True, palette = "bright")
```

```python
# Swarm Plot.
plt.figure(figsize=(28, 8))
sns.swarmplot(data = company, x = "Founded", y = "Number of employees", dodge = True, palette = "bright")
```

```python
# Box Plot.
plt.figure(figsize=(28, 8))
sns.boxplot(data=company,
            x="Founded",
            y="Number of employees",
            dodge=True,   # separate boxes of different colors
            width=50,     # width of boxes
            whis=1.5,     # controls the length of whiskers of a box
            palette = "bright")
```

```python
# Violin Plot.
plt.figure(figsize=(28, 8))
sns.violinplot(data = company,
            x = "Founded",
            y = "Number of employees",
            dodge = True,   # separate plots of different colors
            width = 50,    # width of plots
            palette = "bright",)
```

```python
# Joint Plot.
sns.jointplot(data = company,
              x = "Founded",
              y = "Number of employees",
              kind = "scatter",   # plot kind
              height = 6.5,)      # plot height
```
